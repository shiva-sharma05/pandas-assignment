## Q1. Find Missing Values

Input:
df = pd.DataFrame({
    "Name":["Aman","Riya",None],
    "Age":[21,np.nan,25]
})

Output:
    Name    Age
0  False  False
1  False   True

## Q13. Mini Data Cleaning Task

Input:
data = {
    "Name":["Aman","Raj","Aman",None],
    "Age":[20,None,20,25],
    "City":["Bhopal","Delhi","Bhopal","Indore"]
}

Task:
1. Fill missing Age with mean.
2. Remove duplicate rows.
3. Display final DataFrame.

```python
data = { "Name":["Aman","Raj","Aman",None], "Age":[20,None,20,25], "City":["Bhopal","Delhi","Bhopal","Indore"] }
df = pd.DataFrame(data)
df['Age'] = df['Age'].fillna(df['Age'].mean())
df.drop_duplicates(inplace=True)
df
```
