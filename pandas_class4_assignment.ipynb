# Q1. Select Single Column

## Problem Statement
Create a DataFrame and display only the Name column.

## Sample Input
```python
Name, Age
```

## Sample Output
```python
Only Name column
```


````python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Raj', 'Amit', 'Neha', 'Priya'],
    'Age': [22, 24, 21, 23],
    'Marks': [80, 75, 90, 85],
    'city' : ['bhopal' , 'rewa' , 'indore' , 'bhopal']
})

print(df.columns)
````


# Q2. Select Multiple Columns

## Problem Statement
Display Name and Marks columns from a DataFrame.

## Sample Input
```python
Name, Marks
```

## Sample Output
```python
Selected columns
```


````python
print(df[['Name', 'Marks']])
````


# Q3. Filter Students

## Problem Statement
Display students with Marks greater than 80.

## Sample Input
```python
Marks > 80
```

## Sample Output
```python
Filtered rows
```


````python
df['Marks'][df['Marks'] > 80]
````


# Q4. Filter by City
